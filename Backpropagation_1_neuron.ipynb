```python
import numpy as np
```

```python
np.random.seed(0)
x = np.array([2 ,3])
y = np.array([1])
w = np.random.randn(x.shape[0])
b= np.zeros(1)
```

```python
def neuron(x , w , b):
    return np.dot(x ,w.T) + b
y_pred = neuron(x , w , b)
y_pred
```

```python
def relu(y):
    return np.maximum(0 , y)
y_pred = relu(y_pred)
y_pred
```

```python
def loss(y_pred , y):
    return (y_pred - y)**2 # can also have mean and all shit but only one neuron hai :)
loss_r = loss(y_pred , y)
loss_r
```

```python
# now wht we need to do is reduce this loss :)
lr = 0.001
iteration = 1000
def d_relu(x) :
    if x > 0 :
        return 1 
    else:
        return 0
# weights and biases toh wrt ho rahe hai i guesss so no issue i guess
def  dl_dw_db(x , y , y_pred):
    dl_dw =  2*(y_pred  - y )*x
    dl_db = 2*(y_pred  - y )
    return dl_dw , dl_db
dl_dw , dl_db = dl_dw_db(x , y ,y_pred)
```

##### Penchoo saare functions ready hai bus updating and printing saving ka cheez design karna hai 

```python
loss_min = 9999 # initial value 
for i in range(100000):
    loss_r = loss(y_pred , y)
    w = w - lr*dl_dw*d_relu(y_pred)
    b = b - lr*dl_db*d_relu(y_pred)
    y_pred = neuron(x , w , b)
    dl_dw , dl_db = dl_dw_db(x , y ,y_pred)

    if loss_min > loss_r :
        loss_min = loss_r
        print(loss_min , "  iteration  :" , i  )
        opt_w = w 
        opt_b = b
```
