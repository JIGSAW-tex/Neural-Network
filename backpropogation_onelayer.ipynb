```python
import numpy as np 
```

```python
#What are we designing is 3 neurons and  4 features to make the loss = 1
np.random.seed(0) # random seed 
x = np.array([3 , -4 , 2 , 1]) # 4 feature , 1 sample
y = np.array([5]) # target to be achieved
w = np.random.randn(x.shape[0] , 3 )  # initializing weights
b = np.zeros(3) # no. of biases 
w
```

```python
def neurons(x , w , b):
    y_pred = np.dot(x , w) + b 
    return y_pred
def relu(y_pred):
    return np.maximum(0 , y_pred)
def total(y_pred):
    return np.sum(y_pred)
def loss(y , y_pred):
    return (y-y_pred)**2
def dl_dw_db(x , y , y_pred):
    dl_dw = 2*(y_pred - y)*x
    dl_db = 2*(y_pred - y)
    dl_dw = np.expand_dims(dl_dw , axis= 1)
    return dl_dw , dl_db
def d_relu(y_pred):
    if y_pred > 0 :
        return 1 
    else:
        return 0
```

```python
lr = 0.001
loss_min = 9999
for i in range(10000):
    y_pred = neurons(x , w , b)
    y_pred = relu(y_pred)
    y_pred = total(y_pred)
    loss_r  = loss(y , y_pred)
    dl_dw , dl_db = dl_dw_db(x , y , y_pred)
    w = w - lr*(dl_dw*d_relu(y_pred))
    b = b - lr*(dl_db*d_relu(y_pred))

    if loss_min > loss_r :
        loss_min = loss_r
        w_opt = w 
        b_opt = b 
        print(f"Iteration : {i}  loss : {loss_min}")
```
